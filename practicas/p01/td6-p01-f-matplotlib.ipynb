# **Matplotlib**

Por convención, el alias elegido para Matplotlib (concretamente, para `matplotlib.pyplot`) es `plt`.

```python
import matplotlib.pyplot as plt
```

Tanto en esta *notebook* como en las correspondientes a seaborn y Plotly, a partir de una misma base de datos, vamos a generar una misma figura y, así, poder comparar estas tres librerías.

La base de datos que vamos a usar es sobre viajes en taxi y se encuentra disponible en la misma librería seaborn.

```python
import seaborn as sns # Por convención, el alias elegido para seaborn es sns.
datos = sns.load_dataset(name = 'taxis')
```

```python
datos.head()
```

```python
datos.tail()
```

Mediante la figura que vamos a hacer, queremos ver si a mayor distancia, mayor es la tarifa.

```python
plt.scatter(data = datos, x = 'distance', y = 'fare')
plt.show()
```

Ahora, mejoremos un poco el aspecto de esta figura.

```python
x = datos['distance']
y = datos['fare']

import numpy as np
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.scatter(x = x, y = y, c = 'black', alpha = 0.2)
plt.plot(x, p(x), 'b--')
plt.title('¿A mayor distancia, mayor es la tarifa?')
plt.xlabel('Distancia (mi)') # mi: millas.
plt.ylabel('Tarifa\n(USD)', rotation = 0, labelpad = 25)
plt.show()
```
